```python
import pandas as pd
import ast
occasion = pd.read_csv("occasion_cocktail.csv", index_col=0)
occasion.head()
```

```python
len(occasion)
```

```python
cocktail_dict = {}
cocktail_list = []

for i in range(len(occasion)):
    event = occasion.iloc[i,0]
    cocktails = ast.literal_eval(occasion.iloc[i,1])
    
    for cocktail in cocktails:
        cocktail = cocktail.replace('®',"") # 이름에 기호 빼기
        
        if cocktail not in cocktail_list:
            cocktail_list.append(cocktail)
            cocktail_dict[cocktail] = [event]
        else:
            cocktail_event = cocktail_dict[cocktail]
            cocktail_event.append(event)
            cocktail_dict[cocktail] = cocktail_event
            
cocktail_dict           
```

```python
data_csv = {'Cocktails': list(cocktail_dict.keys()),
        'Occasion': list(cocktail_dict.values())}

df = pd.DataFrame(data_csv)
df
```

```python
df.columns = ['name', 'occasion']
df
```

```python
df.to_csv('occasion.csv')
```
